## HaarCascade Method with OpenCV

Test 1

```python
import cv2
import matplotlib.pyplot as plt

classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
img = cv2.imread('test4.jpg')
faces = classifier.detectMultiScale(img)# result
#to draw faces on image
for result in faces:
    x, y, w, h = result
    x1, y1 = x + w, y + h
    cv2.rectangle(img, (x, y), (x1, y1), (0, 0, 255), 2)
```

```python
plt.imshow(img)
plt.title('Test 1: Haarcascade OpenCV')
plt.show()
plt.imsave('Test 1: Haarcascade OpenCV.png', img)
```

```python
cf = 2
tf = 4
if(cf<tf):
    acc1 = cf/tf
else:
    acc1 = cf/(tf+cf)
    
print(acc1)
```

# Dlib 

Test 1

```python
import dlib
import cv2
import matplotlib.pyplot as plt

detector = dlib.get_frontal_face_detector()
img = cv2.imread('test4.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = detector(gray, 1) # result
#to draw faces on image
for result in faces:
    x = result.left()
    y = result.top()
    x1 = result.right()
    y1 = result.bottom()
    cv2.rectangle(img, (x, y), (x1, y1), (0, 0, 255), 2)
```

# MTCNN 

Test 1

```python
import cv2
from mtcnn.mtcnn import MTCNN
detector = MTCNN()
img = cv2.imread('test4.jpg')
faces = detector.detect_faces(img)# result
#to draw faces on image
for result in faces:
    x, y, w, h = result['box']
    x1, y1 = x + w, y + h
    cv2.rectangle(img, (x, y), (x1, y1), (0, 0, 255), 2)
```

```python
plt.imshow(img)
plt.title('Test 1: MTCNN')
plt.show()
plt.imsave('Test 1: MTCNN.png', img)
```
